<h1> Naive Bayes Model on Kaggle Clement Fake News Dataset </h1>
<h3> The fake news dataset is publically available <a href="https://www.kaggle.com/clmentbisaillon/fake-and-real-news-dataset">here</a></h3>

```python
import util
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import accuracy_score, plot_confusion_matrix
import naive_bayes as nb
```

```python
folder_path_clement = '../../datasets/kaggle_clement/split_files/'
train_file = 'train.csv'
test_file = 'test.csv'

train_X, train_Y = util.load_dataset(folder_path_clement, train_file)
test_X, test_Y = util.load_dataset(folder_path_clement, test_file)
```

<h2> Base Model </h2>

```python
base_model = nb.NaiveBayesModel()
base_model.fit(train_X, train_Y)
```

```python
util.visualize_confusion_matrix(base_model,test_X, test_Y,"naive_bayes_base_clement_confusion_matrix")
```

```python
base_cv_results = cross_validate(base_model, train_X, train_Y, cv=KFold(5))
```

```python
util.plot_learning_curve(base_model, "naive_bayes_base_clement_learning_curve", train_X,train_Y, cv=KFold(5), n_jobs=4)
```
